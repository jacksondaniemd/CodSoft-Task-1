# To-Do List

```python
import json

def load_tasks():
    try:
        with open("tasks.json", "r") as file:
            tasks = json.load(file)
    except FileNotFoundError:
        tasks = []
    return tasks

def save_tasks(tasks):
    with open("tasks.json", "w") as file:
        json.dump(tasks, file)

def display_tasks(tasks):
    if tasks:
        print("Your To-Do List:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
    else:
        print("Your to-do list is empty.")

def add_task(tasks, new_task):
    tasks.append(new_task)
    save_tasks(tasks)
    print("Task added successfully!")

def update_task(tasks, index, updated_task):
    tasks[index - 1] = updated_task
    save_tasks(tasks)
    print("Task updated successfully!")

def remove_task(tasks, index):
    del tasks[index - 1]
    save_tasks(tasks)
    print("Task removed successfully!")

def main():
    tasks = load_tasks()
    
    while True:
        print("\n1. View tasks\n2. Add task\n3. Update task\n4. Remove task\n5. Quit")
        choice = input("Enter your choice: ")
        
        if choice == "1":
            display_tasks(tasks)
        elif choice == "2":
            new_task = input("Enter new task: ")
            add_task(tasks, new_task)
        elif choice == "3":
            display_tasks(tasks)
            index = int(input("Enter task number to update: "))
            updated_task = input("Enter updated task: ")
            update_task(tasks, index, updated_task)
        elif choice == "4":
            display_tasks(tasks)
            index = int(input("Enter task number to remove: "))
            remove_task(tasks, index)
        elif choice == "5":
            print("Exiting program...")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
